```python
with open("fit_params.json",'r') as f:
    fit_params = json.load(f)
```

```python
a = fit_params["a"]
lam = fit_params["lambda"]
c = fit_params["c"]
```

```python
x = np.linspace(0,3564,3564)*25e-9
corr = a*np.exp(-lam*x*1e6) + c
```

```python
plt.plot(corr[0:])
plt.yscale("log")
```

# Correction factors calculated slightly different than usual

```python
def remove_albedo(data,fit,col_idx):
    """
    Remove albedo from data starting from col_idx
    Args:
    data: numpy array - number of hits per bunch crossing
    fit: exponential albedo fit
    col_idx: index of colliding bunch
    """
    reduced = np.zeros(3564)
    A = data[col_idx] #Get number of hits in colliding bunch
    idx = np.array(range(col_idx, 3564+col_idx),dtype = "int")
    idx = idx % 3564
    reduced[idx] = data[idx] -  A*fit
    return reduced
```

```python
reduced = np.array(ch_dict["1"])
for c in tqdm(colliding_bx):
    reduced = remove_albedo(reduced,corr,c)
```
